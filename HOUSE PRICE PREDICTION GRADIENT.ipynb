```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("C:\\Users\\Hp\\Downloads\\train (1).csv")
df
```

```python
df.info()
```

```python
df.drop(df.columns[[6,57,72,73,74]], axis=1, inplace=True)
```

```python
categorical_features = df.select_dtypes(include = ["object","bool"]).columns
numerical_features = df.select_dtypes(include = ["int64","float64"]).columns

```

```python
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].median())
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df[categorical_features] = encoder.fit_transform(df[categorical_features])
```

```python
#df[categorical_features] = df[categorical_features].fillna(method="ffill")
df[categorical_features] = df[categorical_features].fillna(0)
```

```python
df.info()
```

```python
df
```

```python
x=df.drop("SalePrice",axis=1)
y=df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
   
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model= GradientBoostingRegressor()
gb_model.fit(x_train,y_train)


gb_model.score(x_train,y_train)
```

```python
gb_model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(gb_model,x,y,cv=5)
print("cross validation score:",scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
estimator=GradientBoostingRegressor()
param_grid={"n_estimators":[21,22,23,24,25],"learning_rate":[0.1,0.2,0.3,0.5,0.8,1]}
grid=GridSearchCV(estimator,param_grid,cv=5,scoring="neg_mean_squared_error")
grid.fit(x_train,y_train)
print(grid.best_params_)
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
pd.DataFrame(index=x.columns,data=grid.best_estimator_.feature_importances_,columns=["Feature Importance"])
```

```python
feats=pd.DataFrame(data=grid.best_estimator_.feature_importances_,index=x.columns,columns=["importance"])
important_features=feats[feats["importance"]>0.01]
important_features
```

```python
important_features_list=feats[feats["importance"]>0.01].index.to_list()
important_features_list
```

```python
x=x[important_features_list]
y=df["SalePrice"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

gb_bhp=GradientBoostingRegressor(n_estimators=23,learning_rate=0.3)
gb_bhp.fit(x_train,y_train)
```

```python
gb_bhp.score(x_train,y_train)
```

```python
gb_bhp.score(x_test,y_test)
```

```python
scores=cross_val_score(gb_bhp,x,y,cv=5)
print("cross val score:",scores.mean())
```
