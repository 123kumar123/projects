```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("C:\\Users\\Hp\\Downloads\\kidney_disease.csv")
df
```

```python
df["classification"].unique()
```

```python
df['classification'] =df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df.drop_duplicates()
```

```python
df.drop(["id"],axis=1,inplace=True)
```

```python
df['rc']=pd.to_numeric(df['rc'],errors='coerce')
```

```python
df['pcv']=pd.to_numeric(df['pcv'],errors='coerce')
```

```python
df['wc']=pd.to_numeric(df['wc'],errors='coerce')
```

```python
categorical_features = df.select_dtypes(include = ["object","bool"]).columns
numerical_features = df.select_dtypes(include = ["int64","float64"]).columns

```

```python
df[categorical_features]=df[categorical_features].astype("str")
```

```python
df[numerical_features]=df[numerical_features].replace(np.nan,0)
```

```python
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].median())
```

```python
df[categorical_features]=df[categorical_features].replace(np.nan,0)
```

```python
df[categorical_features] = df[categorical_features].fillna(df[categorical_features].mode())
```

```python
df.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df[categorical_features] = encoder.fit_transform(df[categorical_features])
```

```python
df
```

```python
x=df.drop("classification",axis=1)
y=df["classification"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model= GradientBoostingClassifier()
gb_model.fit(x_train,y_train)

y_pred_train=gb_model.predict(x_train)
y_pred_test=gb_model.predict(x_test)

from sklearn.metrics import accuracy_score
print("train_accuracy:",accuracy_score(y_train,y_pred_train))
print("test_accuracy:",accuracy_score(y_test,y_pred_test))


from sklearn.model_selection import cross_val_score
scores=cross_val_score(gb_model,x,y,cv=5)
print("cross validation score:",scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
estimator=GradientBoostingClassifier()
param_grid={"n_estimators":[5,10,15,20],"learning_rate":[0.1,0.2,0.3,0.5,0.8,1]}
grid=GridSearchCV(estimator,param_grid,cv=5,scoring="accuracy")
grid.fit(x_train,y_train)
print(grid.best_params_)
grid.best_estimator_
```

```python
grid.best_estimator_.feature_importances_
```

```python
feats=pd.DataFrame(data=grid.best_estimator_.feature_importances_,index=x.columns,columns=["importance"])
important_features=feats[feats["importance"]>0.01]
important_features
```

```python
important_features_list=feats[feats["importance"]>0.01].index.to_list()
important_features_list
```

```python
x=x[important_features_list]
y=df["classification"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)

gb_bhp=GradientBoostingClassifier(n_estimators=15,learning_rate=0.8)
gb_bhp.fit(x_train,y_train)

ypred_train=gb_bhp.predict(x_train)
predictions=gb_bhp.predict(x_test)

from sklearn.metrics import accuracy_score
print("train accuracy:",accuracy_score(y_train,ypred_train))
print("test accuracy:",accuracy_score(y_test,predictions))
scores=cross_val_score(gb_bhp,x,y,cv=5,)
print("cross validation score:",scores.mean())
```
