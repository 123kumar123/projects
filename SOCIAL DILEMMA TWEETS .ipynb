```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Hp\\Downloads\\TheSocialDilemma.csv (1)\\TheSocialDilemma.csv")
df
```

```python
df.drop(['user_name', 'user_location','user_description', 'user_created',"user_followers","user_friends","user_favourites","user_verified","date","hashtags","source","is_retweet" ], axis=1, inplace=True)
df
```

```python
df.drop(index=df.index[-10000:],inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(len(df)):
    rp=re.sub("[^a-zA-Z]"," ",df["text"][i])
    rp=rp.lower()
    rp=rp.split()
    rp=[ps.stem(word) for word in rp if not word in set(stopwords.words("english"))]
    rp=" ".join(rp)
    corpus.append(rp)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100)
x=cv.fit_transform(corpus).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Sentiment"]=le.fit_transform(df["Sentiment"])
y=df["Sentiment"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model= GradientBoostingClassifier()
gb_model.fit(x_train,y_train)

y_pred_train=gb_model.predict(x_train)
y_pred_test=gb_model.predict(x_test)

from sklearn.metrics import accuracy_score
print("train_accuracy:",accuracy_score(y_train,y_pred_train))
print("test_accuracy:",accuracy_score(y_test,y_pred_test))


from sklearn.model_selection import cross_val_score
scores=cross_val_score(gb_model,x,y,cv=5)
print("cross validation score:",scores.mean())
```
